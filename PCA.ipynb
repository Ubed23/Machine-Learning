```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

```

```python
uni=pd.read_csv("/content/Universities.csv")
uni
```

```python
univ=uni.iloc[:,1:]
univ
```

```python
sc=StandardScaler()
uni_normal=sc.fit_transform(univ)
uni_normal
```

```python
pca=PCA()
pca_components=pca.fit_transform(uni_normal)
```

```python
# the ammount of varience that each PCA explain is
pca.explained_variance_# 1st column contains 4.8042 varience.. upto 6th column 0.002555 varience
```

```python
# in percentage- the ammount of varience that each PCA expains is
var=pca.explained_variance_ratio_
var
```

```python
# cumulative varience
var1=np.cumsum(np.round(var,4)*100)
var1
# 76  76+13=89  89+4=93 and so on
```

```python
pca.components_

# linear combination eqn constant values are taken from above array o/p
# PCA1= 0.45*SAT - 0.427*Top10 + 0.42*Accept + 0.39SFRatio- 0.36*Expenses - 0.37*GardRate

# thus all 6 principal components can be calculated using above 6 values
```

```python
# varience plot for PCA component obtained
plt.plot(var1,color="red")
# var1 is cumulative percentage on y axis and index on x axis
# looking at the graph you can decide how much percentage you want and accordingly go for that much column nos
```
